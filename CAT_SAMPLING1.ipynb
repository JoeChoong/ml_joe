```python
def randomize_answers(sample):
    s = copy.deepcopy(sample)

    for cat in s["categorys"]:
        for q in cat["questionItems"]:
            answers = q["answerItems"]

            # if multiple answers exist, randomly select 1
            if len(answers) > 1:
                chosen = random.choice(answers)
                q["answerItems"] = [chosen]

    return s
```

```python
samples = []
for i in range(5):
    new_sample = randomize_answers(base_sample[0])
    new_sample["id"] = f"ab{i+1}"
    samples.append(new_sample)

print(samples)
```
